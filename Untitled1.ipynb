```python
import pandas as pd
```

```python
#Users
name_tags = ['UserID', 'Gender','Age','Occupation','Zip-code']
usersID= pd.read_table("C:\\Users\\79146\\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\users.dat", sep='::', engine='python',names=name_tags)
usersID.head()
```

```python
#ratings
ratings_tags = ['UserID','MovieID','Rating','Timestamp']
ratings = pd.read_table("C:\\Users\\79146\\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\ratings.dat", sep = '::', engine = 'python',names = ratings_tags )
ratings.head()
```

```python
#movies
movies_tags = ['MovieID','Title','Genres']
movies = pd.read_table("C:\\Users\\79146\\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\movies.dat", sep = '::', engine = 'python',names = movies_tags )
movies.head()
```

```python
#DF=ratings+Users+movies
df = pd.merge(ratings,usersID)
df_main = pd.merge(df,movies)
df_main.head()
```

```python
#сортировка во гендеру 
Stat_gender = df_main.pivot_table("Rating", index = 'Title', columns = "Gender", aggfunc='mean')
Stat_gender.head()
```

```python
#сортировка по возрасту 
Age_stat = df_main.pivot_table("Rating", index = 'Title', columns = "Age",  )
Age_stat.head()
```

```python
#Женшины по убыванию
GendStat = Stat_gender.pivot_table("F", index = 'Title')
GendStat= GendStat.sort_values("F", ascending = False )
GendStat.head()
```

```python
#Старшая группа по возростанию
Old_Stat = Age_stat.pivot_table(56, index = 'Title')
Old_Stat= Old_Stat.sort_values(56, ascending = True )
Old_Stat.head()
```

```python
#groupby()
df_Title = df_main.groupby(["Title"])["Rating"].sum()
df_Title[df_Title >= 250]
```

```python
#Женщины>Мужчины
Stat_gender_2 =  Stat_gender.sort_values("F", ascending = False,)
Stat_gender_2.dropna(inplace=True)
D=Stat_gender_2['F']-Stat_gender_2['M']
DR=D.sort_values(ascending=False)
DR.head(15)
```

```python
#Женщины<Мужчины

DR=D.sort_values(ascending=True)
DR.head(15)
```
